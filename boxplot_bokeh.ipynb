## Interactive Boxplot of Duration of Terminated Recalled Food
## (Top 3 Problems)
#### Note: Duration means the recalled period between initialization and termination.

```python
import pandas as pd
from bokeh.charts import BoxPlot
from bokeh.io import output_notebook, show

df = pd.read_csv("Duration_FoodRecall.csv",encoding='latin1')
df.head()
```

```python
#output_file("boxplot.html")
output_notebook()
```

```python
p = BoxPlot(df, values='duration (month)', 
            label=['problem type','year'], 
            color='problem type',
            legend=False , 
            title="Boxplot of Duration of Terminated Recalled Food (Top 3 Problems)", 
            tooltips=[
                ('Problem Type', '@problem type'),
                ('Year', '@year')])
show(p)
```
